```sql
--PERSISTED OBJ

CREATE VIEW vw_emp as SELECT * FROM employees3 where dept = 'Sales';

SELECT * FROM vw_emp
```

```sql
-- SESSION SCOPED VIEW (Notebook Only, becomes unavaible if the notebook is detached from the cluster or cluster restarts)
CREATE OR REPLACE TEMP VIEW vw_tmp_emp as SELECT * FROM employees3 where dept = 'Sales';

SELECT * FROM vw_tmp_emp
```

```sql
-- CLUSTER SCOPED VIEW (All the notebooks that are using the cluster can access it)
-- NOT SUPPORTED FOR SERVERLESS COMPUTE
CREATE OR REPLACE  GLOBAL TEMP VIEW vw_cluster_emp as SELECT * FROM employees3 where dept = 'Sales';

SELECT * FROM vw_cluster_emp;
```
